```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# 1. Generate random data
np.random.seed(42)  # for reproducibility

n_samples = 300
# Let’s create some correlation between x and y
x = np.random.normal(loc=0.0, scale=1.0, size=n_samples)
y = 2*x + np.random.normal(loc=0.0, scale=1.0, size=n_samples)  # y is somewhat correlated with x

# 2. Create a DataFrame
df = pd.DataFrame({'Feature1': x, 'Feature2': y})

df.head()
df
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(df['Feature1'], df['Feature2'], alpha=0.5)
plt.title("Original Data")
plt.xlabel("Feature1")
plt.ylabel("Feature2")
plt.grid(True)
plt.show()
```

```python
pca = PCA(n_components=2)
pca_data = pca.fit_transform(df)
pca_data.shape
```

```python
pca_data
```

```python
plt.figure(figsize=(10, 4))
plt.scatter(pca_data[:, 0], pca_data[:, 1], alpha=0.5)
plt.title("Data After PCA Transformation")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.grid(True)
plt.show()
```
