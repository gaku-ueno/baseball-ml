# Code to prepare datasets for analysis

```python
import pandas as pd
import numpy as np
import seaborn as sns

from pybaseball import batting_stats
from pybaseball import statcast_batter_expected_stats, statcast_batter_exitvelo_barrels
```

```python
qualifying_threshold = 502

fg_bat_data = batting_stats(2015, 2024, qual = qualifying_threshold)
fg_bat_data
```

```python
fg_bat_data.to_csv('../data/fangraphs_player_data.csv', index = False)
```

```python
exp_bat_data = pd.DataFrame()

for year in range(2015, 2025):
    exp_bat_yearly = statcast_batter_expected_stats(year)
    exp_bat_yearly = exp_bat_yearly[exp_bat_yearly['pa'] > qualifying_threshold]

    exp_bat_data = pd.concat([exp_bat_data, exp_bat_yearly], ignore_index=True)

exp_bat_data
```

```python
exp_bat_data.to_csv('../data/statcast_expected_batting_data', index = False)
```

```python
exit_velo_barrels_data = pd.DataFrame()

for year in range(2015, 2025):
    exit_velo_barrels_data_yearly = statcast_batter_exitvelo_barrels(year)
    exit_velo_barrels_data_yearly['year'] = year
    exit_velo_barrels_data = pd.concat([exit_velo_barrels_data, exit_velo_barrels_data_yearly], ignore_index=True)

exit_velo_barrels_data
```

```python
exit_velo_barrels_data.to_csv('../data/statcast_exvelo_barr_data', index = False)
```
