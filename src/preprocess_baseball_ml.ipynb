# Data Preprocessing

```python
import pandas as pd
import numpy as np
```

```python
raw_player_data = pd.read_csv('../data/fangraphs_data_simplified_cleaned')
raw_player_data
```

```python
raw_player_data.isna().sum()
```

```python
cols_reorder = [col for col in raw_player_data.columns if col != 'wRC+']
cols_reorder.append('wRC+')
raw_player_data = raw_player_data[cols_reorder]
raw_player_data
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

#scaling everything to be between 0 and 1
raw_player_data.iloc[:, :-1] = scaler.fit_transform(raw_player_data.iloc[:, :-1])
raw_player_data
```

```python
raw_player_data.to_csv('../data/fangraphs_data_simple_preprocessed', index = False)
```
